```python
# as there are other values like '16' or '5/6' (I guess for two chords in a measure) I will divide them too

import re
cp_list = []

for row in songs['cp']:
    new = re.findall('\d', row)
    s = ","
    cp_list.append(s.join(new))

bigrams = [b for l in cp_list for b in zip(l.split(",")[:-1], l.split(",")[1:])]
print(bigrams)
```

```python
bigrams_np = np.array([np.array(xi) for xi in bigrams])
```

```python
len(bigrams_np[bigrams_np[:,0] == '1'][bigrams_np[bigrams_np[:,0] == '1'][:,1] == '5'])
```

```python
dict_keys = ["1","2","3","4","5","6","7"]

bigram_matrix = []

for first_key in dict_keys:
    row_list = []
    for second_key in dict_keys:
        row_list.append(len(bigrams_np[bigrams_np[:,0] == first_key][bigrams_np[bigrams_np[:,0] == first_key][:,1] == second_key]))
    bigram_matrix.append(row_list)
```

```python
def print_bigram_matrix(mat):
    print('', end = '\t')
    for key in range(len(dict_keys)):
        print(dict_keys[key], end = '\t')
    print('Second chord')

    for row in range(len(mat)):
        print(dict_keys[row], end = '\t')
        for column in range(len(mat[0])):
            if (isinstance(mat[row][column], int)):
                print(mat[row][column], end = '\t')
            else:
                print("{:.4f}".format(mat[row][column]), end = '\t')
        print('\n')
    print('First chord')
```

```python
print_bigram_matrix(bigram_matrix)
```

```python
bigram_matrix[0][4] / sum(bigram_matrix[0])
```

```python
import copy
norm_bigram_matrix = copy.deepcopy(bigram_matrix)
for i in range(len(bigram_matrix)):
    for j in range(len(bigram_matrix[i])):
        if (sum(bigram_matrix[i]) == 0):
            norm_bigram_matrix[i][j] = 0.0
        else:
            norm_bigram_matrix[i][j] = bigram_matrix[i][j] / sum(bigram_matrix[i])
```

```python
print_bigram_matrix(norm_bigram_matrix)
```

```python
import csv

with open("coocurrence.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerows(norm_bigram_matrix)
```

```python
norm_bigram_matrix = list()
file_name = "coocurrence.csv"
with open(file_name, 'r') as f:
    reader = csv.reader(f)
    
    for row in reader:
        norm_bigram_matrix.append([float(num) for num in row])
```

```python
print_bigram_matrix(norm_bigram_matrix)
```

```python
for row in norm_bigram_matrix:
    if sum(row) == 0:
        row[0] = 0.5
        row[5] = 0.5
```
